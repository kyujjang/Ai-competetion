# time 컬럼 생성(datetime)

```python
from datetime import datetime

bids["time"]/=1000
bids["abs_time"]= bids["time"]
bids['time'] = bids["time"].apply(lambda x: datetime.utcfromtimestamp(x / 1e4))
bids
```

```python
bids["bidder_id"].nunique()
```

# 구매횟수 컬럼

```python
bids2= pd.DataFrame()

bids2= bids.groupby("bidder_id")[["auction","device","country","merchandise","url"]].nunique()

bids2["max_time"]= bids.groupby("bidder_id")["abs_time"].max()
bids2["min_time"]= bids.groupby("bidder_id")["abs_time"].min()
bids2["abs_mean_time"]= bids.groupby("bidder_id")["abs_time"].mean()
bids2['mean_time'] = bids2["abs_mean_time"].apply(lambda x: datetime.utcfromtimestamp(x / 1e4))

bids2["bid_id"]= bids.groupby("bidder_id")["bid_id"].nunique()

bids2["gap_buy"]= (bids2["max_time"]- bids2["min_time"])/bids2["bid_id"]

bids2
```

```python
bids2['year'] = bids2['mean_time'].dt.year
bids2['month'] = bids2['mean_time'].dt.month
bids2['day'] = bids2['mean_time'].dt.day
bids2['hour'] = bids2['mean_time'].dt.hour
bids2['min'] = bids2['mean_time'].dt.minute
bids2['second'] = bids2['mean_time'].dt.second
bids2
```

# 모델링

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import StratifiedKFold

skf=StratifiedKFold(n_splits=10, shuffle=True, random_state= 42)
cbr= CatBoostClassifier(iterations= 10000, learning_rate= 0.05,random_state= 42)

result_cbr= 0
best_score_cbr=0

for train_idx, valid_idx in skf.split(train3, train["outcome"]):
    x_train, x_valid= train3.iloc[train_idx], train3.iloc[valid_idx]
    y_train, y_valid= train['outcome'].iloc[train_idx], train['outcome'].iloc[valid_idx]
    
    cbr.fit(x_train, y_train, eval_set=(x_valid, y_valid), early_stopping_rounds=100)
    result_cbr+= cbr.predict_proba(test3)/10
```

```python
from xgboost import XGBClassifier

xgb=XGBClassifier(learning_rate= 0.05, n_estimators= 10000, colsample_bytree= 0.8, subsample= 0.8, random_state= 42, max_depth= 8)
result_xgb= 0

for train_idx, valid_idx in skf.split(train3, train2["outcome"]):
    x_train, x_valid= train3.iloc[train_idx], train3.iloc[valid_idx]
    y_train, y_valid= train['outcome'].iloc[train_idx], train['outcome'].iloc[valid_idx]
    
    xgb.fit(x_train, y_train, eval_set=[(x_valid, y_valid)], early_stopping_rounds=100)
    result_xgb+= xgb.predict_proba(test3)/10
```
